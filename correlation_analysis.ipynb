## 상관관계

```python
# 능력치와의 관계도
df.corr()['overall_rating'].sort_values(ascending=False)
```

```python
train_cols = ['potential', 'crossing', 'finishing',
       'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve',
       'free_kick_accuracy', 'long_passing', 'ball_control', 'acceleration',
       'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power',
       'jumping', 'stamina', 'strength', 'long_shots', 'aggression',
       'interceptions', 'positioning', 'vision', 'penalties', 'marking',
       'standing_tackle', 'sliding_tackle', 'preferred_foot_left',
       'preferred_foot_right']
dfX = df[train_cols]
dfy = df['overall_rating']
```

```python
# 능력치와 독립변수와의 관계(pairplot)
import seaborn as sns
import matplotlib.pyplot as plt

# subplot 을 이용한 여러개 시각화
f, axes = plt.subplots(7,5,figsize=(30,60))


# 10가지 변수와 연봉의 관계를 시각화
sns.regplot(x=dfX["reactions"], y=dfy, line_kws={'color': 'red'},ax= axes[0,0])
sns.regplot(x=dfX["potential"], y=dfy, line_kws={'color': 'red'},ax= axes[0,1])
sns.regplot(x=dfX["ball_control"], y=dfy, line_kws={'color': 'red'},ax= axes[0,2])
sns.regplot(x=dfX["short_passing"], y=dfy, line_kws={'color': 'red'},ax= axes[0,3])
sns.regplot(x=dfX["shot_power"], y=dfy, line_kws={'color': 'red'},ax= axes[0,4])
sns.regplot(x=dfX["long_passing"], y=dfy, line_kws={'color': 'red'},ax= axes[1,0])
sns.regplot(x=dfX["vision"], y=dfy, line_kws={'color': 'red'},ax= axes[1,1])
sns.regplot(x=dfX["dribbling"], y=dfy, line_kws={'color': 'red'},ax= axes[1,2])
sns.regplot(x=dfX["long_shots"], y=dfy, line_kws={'color': 'red'},ax= axes[1,3])
sns.regplot(x=dfX["crossing"], y=dfy, line_kws={'color': 'red'},ax= axes[1,4])
sns.regplot(x=dfX["penalties"], y=dfy, line_kws={'color': 'red'},ax= axes[2,0])
sns.regplot(x=dfX["positioning"], y=dfy, line_kws={'color': 'red'},ax= axes[2,1])
sns.regplot(x=dfX["curve"], y=dfy, line_kws={'color': 'red'},ax= axes[2,2])
sns.regplot(x=dfX["heading_accuracy"], y=dfy, line_kws={'color': 'red'},ax= axes[2,3])
sns.regplot(x=dfX["volleys"], y=dfy, line_kws={'color': 'red'},ax= axes[2,4])
sns.regplot(x=dfX["free_kick_accuracy"], y=dfy, line_kws={'color': 'red'},ax= axes[3,0])
sns.regplot(x=dfX["stamina"], y=dfy, line_kws={'color': 'red'},ax= axes[3,1])
sns.regplot(x=dfX["finishing"], y=dfy, line_kws={'color': 'red'},ax= axes[3,2])
sns.regplot(x=dfX["aggression"], y=dfy, line_kws={'color': 'red'},ax= axes[3,3])
sns.regplot(x=dfX["strength"], y=dfy, line_kws={'color': 'red'},ax= axes[3,4])
sns.regplot(x=dfX["sprint_speed"], y=dfy, line_kws={'color': 'red'},ax= axes[4,0])
sns.regplot(x=dfX["interceptions"], y=dfy, line_kws={'color': 'red'},ax= axes[4,1])
sns.regplot(x=dfX["acceleration"], y=dfy, line_kws={'color': 'red'},ax= axes[4,2])
sns.regplot(x=dfX["agility"], y=dfy, line_kws={'color': 'red'},ax= axes[4,3])
sns.regplot(x=dfX["jumping"], y=dfy, line_kws={'color': 'red'},ax= axes[4,4])
sns.regplot(x=dfX["standing_tackle"], y=dfy, line_kws={'color': 'red'},ax= axes[5,0])
sns.regplot(x=dfX["balance"], y=dfy, line_kws={'color': 'red'},ax= axes[5,1])
sns.regplot(x=dfX["marking"], y=dfy, line_kws={'color': 'red'},ax= axes[5,2])
sns.regplot(x=dfX["sliding_tackle"], y=dfy, line_kws={'color': 'red'},ax= axes[5,3])
sns.regplot(x=dfX["preferred_foot_right"], y=dfy, line_kws={'color': 'red'},ax= axes[5,4])
sns.regplot(x=dfX["preferred_foot_left"], y=dfy, line_kws={'color': 'red'},ax= axes[6,0])

plt.show()
```

```python
# 상관계수의 검점
# 독립변수와 능력치의 상관관계
import numpy as np
from scipy import stats

y = dfy.values
# 3가지 변수를 반복문을 통해 결과값을 분석
for item in dfX.columns:
    print(item)
    X = dfX[item].values
    print ('Covariance : {:.2f}'.format(np.cov(X,y)[0,1])) # 공분산
    print ('Correlation : {:.2f}'.format(stats.pearsonr(X,y)[0])) # 상관분석
    print ('P-value : {:.4f}'.format(stats.pearsonr(X,y)[1])) # P-밸류
    print('\n') 
```
